```python
def check_ordering(sequence, before_list, after_list):
    for i in range(len(sequence)):
        if sequence[i] in before_list:
            # Find all indices where sequence[i] appears in before_list
            indices_of_before = [
                index for index, value in enumerate(before_list) if value == sequence[i]
            ]

            for index_of_before in indices_of_before:
                before_value = after_list[index_of_before]
                if before_value in sequence:
                    max_index = sequence.index(before_value)
                    if i > max_index:
                        return False
    return True


for sequence in update_list:
    update_log.append(check_ordering(sequence, before_values, after_values))

import math


def get_middle_value(sequence):
    sequence_length = len(sequence)
    halfway_point = math.ceil(sequence_length / 2) - 1
    return sequence[halfway_point]


def sort_update_list(update_list_copy, should_skip_list, before_list, after_list):
    for j in range(len(update_list_copy)):
        if should_skip_list[j] == False:
            for i in range(len(update_list_copy[j])):
                if update_list_copy[j][i] in before_list:
                    # Find all indices where sequence[i] appears in before_list
                    indices_of_before = [
                        index
                        for index, value in enumerate(before_list)
                        if value == update_list_copy[j][i]
                    ]
                    for index_of_before in indices_of_before:
                        before_value = after_list[index_of_before]
                        if before_value in update_list_copy[j]:
                            max_index = update_list_copy[j].index(before_value)
                            if i > max_index:
                                # Swap the locations of the 'before' and 'after' values in the sequence
                                before_index = update_list_copy[j].index(
                                    update_list_copy[j][i]
                                )
                                after_index = max_index

                                # Perform the swap
                                (
                                    update_list_copy[j][before_index],
                                    update_list_copy[j][after_index],
                                ) = (
                                    update_list_copy[j][after_index],
                                    update_list_copy[j][before_index],
                                )
                                print(update_list_copy[j])
    return update_list_copy


update_list = sort_update_list(update_list, update_log, before_values, after_values)
for i in range(len(update_list)):
    if update_log[i] == True:
        middle_sum_true += get_middle_value(update_list[i])
    else:
        middle_sum_false += get_middle_value(update_list[i])

print(middle_sum_false)
```
